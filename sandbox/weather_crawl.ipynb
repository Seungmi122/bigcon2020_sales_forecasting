```python
# import libraries
import pandas as pd
import numpy as np
import requests
import re
from bs4 import BeautifulSoup
```

```python
# define crawling function
def crawl(stn, mm, x, y, prefix='https://www.weather.go.kr/weather/climate/past_cal.jsp?'):
    """
    :param stn:int - station number
    :param mm:int(1-12) - month
    :param x&y:int - unique id for each station number
    :return:pd.dataframe - average, highest, lowest temp and precipitation for month
    """
    website = requests.get(prefix+'stn='+str(stn)+'&yy=2019&mm='+str(mm)+'&'+str(x)+'&'+str(y)+'&'+'obs=1')
    whole = website.content
    one = BeautifulSoup(whole, 'html.parser')
    part = one.find_all('td')

    content = []

    for child in part:
        stats = child.get_text('|')
        content.append(stats.split('|'))

    content_real = []
    for a in content:
        if len(a) > 1:
            content_real.append(a)

    aver = []
    high = []
    low = []
    rain = []

    for child in content_real:
        a = child[0].replace('℃',':')
        a = a.split(':')
        aver.append(a[1])
    
        b = child[1].replace('℃',':')
        b = b.split(':')
        high.append(b[1])
        
        c = child[2].replace('℃',':')
        c = c.split(':')
        low.append(c[1])
        
        if child[4]==' - ':
            pass
        else:
            d = child[4].replace('mm',':')
            d = d.split(':')
            rain.append(d[1])

    final = pd.DataFrame({'aver':aver, 'high':high, 'low':low, 'rain':rain})
    final.index = final.index + 1
    return final
```

```python
crawl(stn=99, mm=8, x = 25, y = 14)
```
